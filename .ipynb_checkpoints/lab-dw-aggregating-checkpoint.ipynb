```python
# Importing the file and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
marketing_df=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")
```

```python
marketing_df
```

```python
#changing the column names first
new_columns = [col.replace(' ', '_') for col in marketing_df.columns]

marketing_df.columns = new_columns

marketing_df["Total_Claim_Amount"]
```

```python
#Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.
customer_sorted=marketing_df.groupby("Customer")[["Total_Claim_Amount"]]
customer_sorted = marketing_df.sort_values(by= "Total_Claim_Amount", ascending=False)
customer_sorted
```

```python
#customers claiming more than 1000 and said yes 
rich_customers = customer_sorted[(customer_sorted['Total_Claim_Amount'] > 1000) & (customer_sorted['Response'] == 'Yes')]
rich_customers
```

```python
amount_state=rich_customers.groupby(["Gender","Policy_Type"])[["Total_Claim_Amount"]].mean()
amount_state
```

```python
customers_by_state = marketing_df.groupby('State', as_index = False )[['Customer']].count()
customers_by_state

filtered_states = marketing_df[["Customer"] > 500]

filtered_states
```

```python
#Find the median customer lifetime value by education level and gender.
median_value=marketing_df.groupby(["Education","Gender"])[["Customer_Lifetime_Value"]].median()
median_value
```

```python
#Find the maximum, minimum, and average number of open complaints by state and policy type.
max_open_complaints = marketing_df.groupby(["State","Policy_Type"])[["Number_of_Open_Complaints"]].max()
min_open_complaints = marketing_df.groupby(["State","Policy_Type"])[["Number_of_Open_Complaints"]].min()
avg_open_complaints = marketing_df.groupby(["State","Policy_Type"])[["Number_of_Open_Complaints"]].mean()
avg_open_complaints
```

```python
marketing_df['Effective_To_Date'] = pd.to_datetime(marketing_df["Effective_To_Date"])
marketing_df['Month'] = marketing_df['Effective_To_Date'].dt.month
marketing_df["Month"] = marketing_df["Effective_To_Date"].dt.month

marketing_df

policies_sold=marketing_df.groupby(["State","Effective_To_Date"], as_index= False)[["Number_of_Policies"]].count()

pivot_table = policies_sold.pivot(index='State', columns='Effective_To_Date', values='Number_of_Policies')

pivot_table
```
